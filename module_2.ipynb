#### Removing columns

```python
ufo = read_csv("http://bit.ly/uforeports")
ufo.head()
```

```python
ufo.columns = ufo.columns.str.replace(' ','_')
ufo.head(2)
```

```python
ufo.shape
```

```python
ufo.drop("Colors_Reported",axis = 1,inplace = True)
ufo.head()
```

```python
ufo.drop("City", axis = 1, inplace = True)
ufo.head()
```

```python
ufo.drop(['State', 'Time'],axis = 1, inplace=True)
ufo.head() 
```

```python
ufo.drop([0,1], axis = 0, inplace = True)
ufo.head() #droping Rows..........
```

#### Apply multiple filter criteria 

```python
movie2 = read_csv('http://bit.ly/imdbratings')
movie2.head(2)
```

```python
movie2[movie2.duration >= 200].head(2)
```

```python
movie2[movie2['duration'] >= 200].head(2)
```

```python
#Logical operator...AND or OR   
```

```python
movie2[(movie2.duration >= 200) & (movie2.genre == 'Drama')]
```

```python
movie2[(movie2.duration >= 200) | (movie2.genre == 'Drama')].head()
```

```python
(movie2.duration >= 200) & (movie2.genre == 'Drama')
```

```python
movie2[(movie2.genre == 'Crime') & (movie2.genre == 'Drama') | (movie2.genre == 'Action')].head()
```

```python
movie2[movie2.genre.isin(['Crime', 'Drama','Action'])].head()
```
